```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

# generates a gaussian r.v x
x = stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5) #kernel density plot for plotting cdf curve
plt.show()
```

```python
stats.kstest(x, 'norm')
```

```python
y = np.random.uniform(0, 1, 10000) # continuous uniform dib(0, 1)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```
