```python
print(iris.columns)
```

<h1>PDF And CDF

```python
iris_setosa = iris['petal_length'][iris['species'] == 'setosa']
counts, bin_edges = np.histogram(iris_setosa, bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.show()
```

<h1> Mean and Standard Deviation
    

```python
iris_virginica = iris['petal_length'][iris['species'] == 'virginica']
iris_versicolor = iris['petal_length'][iris['species'] == 'versicolor']

print("Means:")
print(np.mean(iris_setosa))
print(np.mean(iris_virginica))
print(np.mean(iris_versicolor))

print("\nStd-dev:")
print(np.std(iris_setosa))
print(np.std(iris_virginica))
print(np.std(iris_versicolor))
```

<h1> Median

```python
print("Medians: ")
print(np.median(iris_setosa))
print(np.median(iris_virginica))
print(np.median(iris_versicolor))
```

<h1> Percentiles and Quantiles
    

```python
print("90th Percentiles:")
print(np.percentile(iris_setosa, 90))
print(np.percentile(iris_virginica, 90))
print(np.percentile(iris_versicolor, 90))

print("Quantiles")
print(np.percentile(iris_setosa, np.arange(0, 100, 25)))
print(np.percentile(iris_virginica, np.arange(0, 100, 25)))
print(np.percentile(iris_versicolor, np.arange(0, 100, 25)))
```

<h1> Median Absolute Deviation

```python
from statsmodels import robust
print("Median absolute deviation")
print(robust.mad(iris_setosa))
```
